```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/retail_sales_dataset.csv')
```

```python
df.head()
```

```python
print(df.info())
```

```python
print(df.shape)
```

```python
import statistics
```

```python
column_names = list(df)
for column in column_names:
    if df[column].dtype != 'object':
        mean = statistics.mean(df[column])
        print(f"Mean of column '{column}': {mean}")
```

```python
column_names = list(df)
for column in column_names:
    if df[column].dtype != 'object':
        median = statistics.median(df[column])
        print(f"Median of column '{column}': {median}")
```

```python
column_names = list(df)
for column in column_names:
    if df[column].dtype != 'object':
        mode = statistics.mode(df[column])
        print(f"Mode of column '{column}': {mode}")
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
df.groupby("Product Category")["Age"].mean().plot(kind="line")
```

```python
df["Age"].plot.hist(bins=5)
```
